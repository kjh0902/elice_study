matplotlib.pyplot: 다양한 데이터를 시각화해주는 파이썬 모듈이다

```python
## line plot(선 그래프)
## plot(x,y): 각 (x,y)를 직선으로 잇는 그래프를 그린다
# linestyle: 선의 스타일
# marker: 점의 스타일
# color: 선의 색깔
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x = np.arange(15)
y = x**2
ax.plot(
    x,y,
    linestyle=":",
    marker="*",
    color="#524FA1"
)
```

```python
## 한 개의 ax에 여러 선을 나타낼 수 있다
# color는 여러 방식으로 표현할 수 있다
# "r"(red),"green","0.8"(0~1의 회색 scale),"#524FA1"(RGB 16진수)
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x,x,linestyle="-")    #y=x
ax.plot(x,x+2,linestyle="--")   #y=x+2
ax.plot(x,x+4,linestyle="-.")   #y=x+4
ax.plot(x,x+6,linestyle=":")    #y=x+6
```

```python
# color는 여러 표현 방식이 있다
# "r"(red),"green","0.8"(회색 scale),"#524FA1"(RGB 16진수)

# marker는 여러 표현 방식이 있다
# . o v s *
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x,x,color="r",marker=".")    #y=x
ax.plot(x,x+2,color="green",marker="o")   #y=x+2
ax.plot(x,x+4,color="0.8",marker="v")   #y=x+4
ax.plot(x,x+6,color="#524FA1",marker="s")    #y=x+6
ax.plot(x,x+8,color="#AAAAAA",marker="*")    #y=x+8
```

```python
## 축 경계를 조정하는 함수: set_xlim(), set_ylim()
# numpy의 linspace(start,stop,num) 함수: start와 stop 사이를 일정한 간격으로 num개의 요소를 만든다
x = np.linspace(0,np.pi*4,100)
fig, ax = plt.subplots()
ax.plot(x,np.sin(x), marker="*")
ax.set_xlim(-1,13)
ax.set_ylim(-1.5,1.5)
```

```python
## 범례: 선의 label을 설명해주는 별도의 패드
## legend() 함수로 범례를 만든다
x=np.arange(10)
fig,ax=plt.subplots()
ax.plot(x,x,label='y=x')
ax.plot(x,x**2,label='y=x^2')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(loc='upper right',  #패드의 위치
          shadow=True,  #패드의 그림자
          fancybox=True,  #패드의 모서리를 둥글게
          borderpad=2)  #패드의 크기조절
```

```python
## bar plot(막대 그래프)
# bar(x,y) 함수: x축에 y=2*x에 해당하는 bar의 plot을 생성한다
x = np.arange(3)
years = ['2018', '2019', '2025']
values = [100, 400, 900]

fig,ax = plt.subplots(figsize=(12,4))   #figure size: 가로=12, 세로=4
ax.bar(x, values)
ax.set_xticks(x, years)   #set_xticks() 함수: x축의 눈금에 표시한다(0 1 2 -> 2018 2019 2025)
```

```python
## 여러 막대 그래프 쌓기
# np.random.rand(n) 함수: 0~1 사이의 무작위의 수 n개를 생성한다
x = np.random.rand(3)   #파란색 막대 3개의 높이
y = np.random.rand(3)   #주황색 막대 3개의 높이
z = np.random.rand(3)   #초록색 막대 3개의 높이
data = [x,y,z]
fig,ax = plt.subplots()
x_ax = np.arange(3)   #x_ax = [0 1 2]
for i in x_ax:
  ax.bar(x_ax, data[i],
         bottom=np.sum(data[:i], axis=0))   #bottom 인자: 막대를 어디서부터 시작할지 정한다

ax.set_xticks(x_ax)
ax.set_xticklabels(["A","B","C"])
```

```python
## 도수분포표(histogram)
## hist(data) 함수로 히스토그램을 생성한다
# bins는 막대의 개수를 정하는 매개변수다
# 막대의 개수에 따라 구간의 크기가 달라지므로 bins를 적당한 수로 설정한다
fig, ax = plt.subplots()
data = np.random.randn(1000)    #가우시안 표준 정규 분포에서 난수 1000개를 생성한다
ax.hist(data, bins=5)
ax.hist(data, bins=10)
```

```python
## 판다스를 이용한 matplotlib의 그래프 그리기

## 선 그래프 그리기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/sample_data/california_housing_train.csv")
fig, ax = plt.subplots()
ax.plot(np.arange(17000),df['latitude'],label='latitude')
ax.set_xlabel('number')
ax.set_ylabel('latitude')
```

```python
ex = pd.DataFrame({
    'Name':['a','b','c','d','e','f'],
    'Type':['fire','fire','water','water','water','fire'],
    'Attack':[49,62,82,108,52,64],
    'Defense':[49,63,83,71,43,58]
})
ex.to_csv("/content/sample_data/ex.csv")

## 산점도 함수 scatter(x,y)
df = pd.read_csv("/content/sample_data/ex.csv")
fire = df[df['Type'] == 'fire']
water = df[df['Type'] == 'water']
fig, ax = plt.subplots()
ax.scatter(fire['Attack'], fire['Defense'], color="red", label='Fire',marker="*",s=50)
ax.scatter(water['Attack'], water['Defense'], color="blue", label='Water',s=25 )
ax.set_xlabel("Attack")
ax.set_ylabel("Defense")
ax.legend(loc='upper right')
```
